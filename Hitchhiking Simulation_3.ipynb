## Define the distribution of hitchhiking jump sizes $\rho(h)$

+ At the moment it uses $\rho \sim 1/h$
+ Normalized using lower threshold $\rho(h) = h^{-1}/\int_e^1h^{-1} dh$. 
+ This means that the cumulative up to some value $h$ is given by $1-\ln h / \ln e$. 
+ In the dynamics below hitchhiking events occur at rate $\Gamma$ (probability per unit time of drawing a sample from $\rho$)
+ In theory, gamma is defined as the coefficient in front of the $1/h$ term
+ Therefore the Gamma used here ($\Gamma_J$) and the Gamma from Daniel's notes ($\Gamma_D$) are related by: $\Gamma_D = \Gamma_J / \ln e$
+ This means that $\phi = 2 \mu / \Gamma_D = 2 \ln(e)\mu / \Gamma_J$
+ I choose $ln(e) = 5$ (cuts off jumps below 0.7 % or so) meaning $\phi = 10 \mu / \Gamma_J$

```python
# Generate the hitchhiking jump kernel and a function that generates random samples from it

dh=10**(-6)
e=0.001 #1/exp(5.0)
h_list=np.linspace(e, 1.0, 1.0/dh)
cumulative_h={}
for h in h_list:
    cumul=round(1.0-(math.log(h)/math.log(e)), 3)
    cumulative_h[cumul]=h

def h_sample():
    "Stochastic function returning the value of h for hitchhiking event"
    eta=random.random()
    h=cumulative_h[round(eta, 3)]
    return h;

n_sites=2      #number of sites
```

## Useful functions

```python
# Define the Mutate, Recombine, and Hitchhike functions. Number of (neutral) sites chosen too

n_sites=2      #number of sites

def hamdist(str1, str2):
    "Count the # of differences between equal length strings str1 and str2"
    diffs = 0
    for ch1, ch2 in zip(str1, str2):
        if ch1 != ch2:
            diffs=diffs+1
    return diffs
s='{0:0%ib}'% n_sites
genotypes=[s.format(i) for i in range(2**n_sites)] # if change n_sites need to change formatting here.

def weighted_choice(choices):
    total = sum(w for c, w in choices)
    r = random.uniform(0, total)
    upto = 0
    for c, w in choices:
        if upto + w > r:
            return c
        upto += w

M=[]
for genotype1 in genotypes:
    neighbors=[]
    for genotype2 in genotypes:
        if hamdist(genotype1, genotype2)==1:
            neighbors.append(int(genotype2, 2))
    M.append(neighbors)
    
D={}
for i in range(len(genotypes)-1):
    genotype1=genotypes[i]
    index1=int(genotype1, 2)
    for j in range(i+1,len(genotypes)):
        genotype2=genotypes[j]
        index2=int(genotype2, 2)
        for breakpoint in range(1,n_sites):
#             print 'breakpoint=', breakpoint
#             print 'genotype 1 =', genotype1
#             print 'genotype 2 =', genotype2
            genotype3=genotype1[0:breakpoint]+genotype2[breakpoint:len(genotype2)]
            index3=int(genotype3, 2)
#             print 'genotype 3 =', genotype3
            genotype4=genotype2[0:breakpoint]+genotype1[breakpoint:len(genotype1)]
            index4=int(genotype4, 2)
#             print 'genotype 4 =', genotype4
            
            if sorted((index1, index2))!= sorted((index3, index4)):
                if (index1, index2) in D:
                    D[(index1, index2)].append((index3, index4))
                else:
                    D[(index1, index2)]=[(index3, index4)]
#                 print 'added to D'

            
def mutate(f_vect, mu, dt):
    #takes genotype freq vector and returns the change in freq after a fraction, mu, of genotypes mutate
    Df=np.array([0.0 for i in range(2**n_sites)])
    for i in range(len(f_vect)):
        neighbors=M[i]
        df=[]
        for j in neighbors:
            df.append(f_vect[i]*(-mu*dt)+f_vect[j]*mu*dt)
        ddf=sum(df)
        Df[i]=ddf        
    return np.array(Df)

def recombine(f_vect, r, dt):
    #takes genotype freq vector and returns the frequency change after a fraction r of genotypes undergo recombination
    Df=np.array([0.0 for i in range(2**n_sites)]) #a list of all changes in frequencies due to all pair interactions
    for recomb_pair in D:
        index1=recomb_pair[0]
        index2=recomb_pair[1]
        
        formable_pairs=D[recomb_pair]
        for formed_pair in formable_pairs:
            index3=formed_pair[0]
            index4=formed_pair[1]

            df=np.array([0. for i in range(2**n_sites)])
            
            df[index1]=-(r*dt)*f_vect[index1]*f_vect[index2]            
            df[index2]=-(r*dt)*f_vect[index1]*f_vect[index2]
            df[index3]=(r*dt)*f_vect[index1]*f_vect[index2]
            df[index4]=(r*dt)*f_vect[index1]*f_vect[index2]
#             print df
            Df=Df+df  
#     print Df 
    return Df

def hitchhike(f_vect):
    #returns change in freq vector after a single hitchhiking event. Also outputs identity of hh'ing geno
    h=h_sample()
    index_chosen=weighted_choice(list(enumerate(f_vect)))
    Df=np.array([-h*f_vect[i] for i in range(2**n_sites)])
    Df[index_chosen]=(1.0-f_vect[index_chosen])*h
    return [Df, index_chosen]

def log_trans(f, thresh):
    #returns freq transformed by ln(f/(1-f)) with some minimum freq threshold so displayable on log
    log((f+thresh)/(1.0 - f-thresh))
    return log((f+thresh)/(1.0 - f-thresh))
```

## Parameters

+ Outputs `freqs` which is a list (over all the sampled time points) of the the vector containing the frequencies of the genotypes. If there are 2 sites it is a vector of length 4 representing the 00, 01, 10, 11 genotypes.
+ The frequencies are updates every `dt` which we typically set to `1`.
+ The frequencies are *recorded* every `dt_sample` which should be set to roughly the correlation time of frequencies 

```python
# Parameters for the simulated dynamics

dt=1
N=10**(3)        # population size 
mu=10**(-5)     # mutation rate 
r=10**(-2)    # recombination rate
Gamma=10**(-10.)   # rate of hitchhiking  
stochastic=1      # whether to include drift, 1=yes, 0=no

dt_sample=10   # time between sampling normally 100 for long-term or 1 for transients
t_sample=0    #time to begin sampling
run_time=1000000   #500 for transients, round(3.1/mu)  for long-term
number_of_timesteps=int(run_time/dt)
simulations=100

effective_drift_boundary=1 # whether to do effective boundary for drift

print('Is sim stochastic:', stochastic)
print()
print('N=',N)
print('1/u=',1.0/mu)
print('1/Gamma=',1.0/Gamma)
print('1/r=',1.0/r)
print()
phi=2*mu/(Gamma/math.log(1.0/e))
print('phi=', 2*mu/(Gamma/math.log(1.0/e)))
print( 'theta=', N*mu)
chi=1.0/(N*(Gamma/math.log(1.0/e)))
print( 'chi=', 1.0/(N*(Gamma/math.log(1.0/e))))
print()
print('run time of single sim=', run_time)
print('sampling time=', dt_sample)

hh_threshold=phi
```

### The effective boundary

+ In strong drift regime, a fraction $\sim 1-2\theta$ of sites sit on the corners (neither is polymorphic) and a fraction $\sim \theta$ sit on the edges. In the joint distributions we are most interested in stuff in the middle i.e. both sites polymorphic, but this happens with probability $\sim \theta^2$ which is small enough that it can be hard to get enough data.

+ To sample more efficiently what happens in the middle, each time a site hits the boundary, we can choose to push it off again in direction(s) that come from single mutants

## The dynamics

```python
# Run the dynamics 

frequency_trajectories=[]
frequency_hh_jumps=[]

for sim in range(simulations):
    if sim%10==0:
        print('simulation: ',sim)
    
    
    freqs=[]          #list of the frequency trajectories for each genotype [[g1_t0, g1_t1...],[g2_t0, g2_t1...]...] 
    hh_jumps=[] 
    # list of the series of jumps for each genotype. [[g1 jump1, g1 jump2...], [g2...]]
    #the entry for each is a list of [time, start freq, end freq]

    
    # initialize time
    t=0.0
    time_for_next_hh=np.random.exponential(1.0/Gamma)
    
    # initialize freqs
    f=np.array([0.0 for i in range(2**n_sites)]) # vector of genotype frequencies
    f[0]=1.0 
    # print f
    
    for timestep in range(number_of_timesteps):

        t=t+dt
        if t%10**5==0:
            print('t=', t)

        df_mut=mutate(f, mu, dt)

        df_rec=recombine(f, r, dt) 

        if t>time_for_next_hh:
            time_for_next_hh=t+np.random.exponential(1.0/Gamma)
            df_identity=hitchhike(f)
            df_hh=df_identity[0]
            index_hh=df_identity[1]
            time_jump=[t,[list(f),list(f+df_hh)]]
            if df_hh[index_hh]>hh_threshold:
                hh_jumps.append(time_jump)

        else:
            df_hh=np.array([0.0 for i in range(2**n_sites)])

    #     print "***TIMESTEP %i***" %t 
    #     print 'old_f=', f
    #     print 'df_mut=', df_mut
    #     print 'df_rec=', df_rec
    #     print 'df_hh=', df_hh

        new_f1= f + df_mut + df_rec + df_hh    #update the frequencies from everything except drift
    #     print 'new_f1=', new_f1

        if stochastic==1:
            new_f1=np.clip(new_f1, 0.0, 1.0)
            numbers=[np.random.binomial(N, ff) for ff in new_f1] #noise binomial sampling
            total_number=sum(numbers) 
            
            if effective_drift_boundary==1: #implements effective boundary
                #if on one of the corners choose one of the two hamming distance 1 genos to throw in
                if numbers[0]==total_number:
                    if np.random.random()>0.5:
                        numbers[1]=1
                    else:
                        numbers[2]=1
                        
                if numbers[1]==total_number:
                    if np.random.random()>0.5:
                        numbers[0]=1
                    else:
                        numbers[3]=1
                
                if numbers[2]==total_number:
                    if np.random.random()>0.5:
                        numbers[0]=1
                    else:
                        numbers[3]=1
                
                if numbers[3]==total_number:
                    if np.random.random()>0.5:
                        numbers[1]=1
                    else:
                        numbers[2]=1
                        
                if numbers[0]+numbers[1]==total_number:
                    if np.random.random()<float(numbers[0])/total_number:
                        numbers[2]=1
                    else:
                        numbers[3]=1
                if numbers[1]+numbers[3]==total_number:
                    if np.random.random()<float(numbers[1])/total_number:
                        numbers[0]=1
                    else:
                        numbers[2]=1
                if numbers[2]+numbers[3]==total_number:
                    if np.random.random()<float(numbers[2])/total_number:
                        numbers[0]=1
                    else:
                        numbers[1]=1
                if numbers[0]+numbers[2]==total_number:
                    if np.random.random()<float(numbers[0])/total_number:
                        numbers[1]=1
                    else:
                        numbers[3]=1
                        
            total_number=sum(numbers) 

            new_f2=[float(n)/total_number for n in numbers] # normalise frequencies to 1
            
            
        else:
            new_f2=np.clip(new_f1, 0.0, 1.0)

    #     print 'new_f2=', new_f2

        #record the freqs to vector every dt_sample generations
        t_sample=t_sample+dt
        if t_sample>=dt_sample:
    #         print 'appended f=', f
            freqs.append(list(f))
            t_sample=0

        f=new_f2
    
    frequency_trajectories.append(freqs)
    frequency_hh_jumps.append(hh_jumps)
```

# Single site frequency spectra

```python
# extract Site freq data from freq trajectories
def SFS_at_t(freq_trajs, sampling_time, interval_start, interval_end):
    
    site_freqs=[]
    for traj in freq_trajs:
        
        for i, f in enumerate(traj):
            time_of_sample=(i*sampling_time)
            if interval_start<=time_of_sample<=interval_end:
                site_freqs.append(f[1]+f[3])
                site_freqs.append(f[2]+f[3])
    
    return site_freqs
```

## Regime 1

```python
# SFS theory: REGIME 1

def drift_mutation_solution(N, mu, t, c1, c2, p, dp):
    density=((2*N*mu/p)*math.exp(-c2*mu*t)  +  N*mu*(1.0-math.exp(-c2*mu*t))/(p*(1.0-p)))*math.exp(-c1*N*p/t)*dp
    return density

def drift_mutation_solution_log(N, mu, t, c1, c2, y, dy):
    
    p=10**y/(1.0+10**y)
    dp=math.log(10)*p*(1-p)*dy
    
    density=((2*N*mu/p)*math.exp(-c2*mu*t)  +  N*mu*(1.0-math.exp(-c2*mu*t))/(p*(1.0-p)))  *   math.exp(-c1*N*p/t)*dp
    return density

def drift_mutation_solution_start(N, mu, t, c1, c2, p, dp):
    
    if t<=N:
        sol=(1.0-c1*N*mu*math.log(t))
    elif t>N:
        sol=0.5*(1.0-c1*N*mu*math.log(N))*(1.0-math.exp(-c2*mu*t)) + (1.0-c1*N*mu*math.log(N))*(math.exp(-c2*mu*t))
    return sol

def drift_mutation_solution_end(N, mu, t, c1,c2, p, dp):
    
    if t<=N:
        sol=0.0
    elif t>N:
        sol=0.5*(1.0-c1*N*mu*math.log(N))*(1.0-math.exp(-c2*mu*t))
    return sol
```

```python
# SFS theory: REGIME 2

def drift_draft_solution_edge_log(N, mu, Gamma, c1, c3, y, dy):
    
    p=10**y/(1.0+10**y)
    dp=math.log(10)*p*(1-p)*dy
    
    density=N*mu/(p*(1.0-p))*math.exp(-N*Gamma*c1*c3*(p*(1-p)))*dp
    return density

def drift_draft_solution_middle_log(N, mu, Gamma, c1, c3, y, dy):
    
    p=10**y/(1.0+10**y)
    dp=math.log(10)*p*(1-p)*dy
    
    ss=-N*mu*dp*math.log(1.0-(1.0/(c3*Gamma*N))/(p*(1-p)))
    return ss

def transient_drift_draft_solution_edge_log(N, mu, Gamma, c1, c3, t, y, dy):
    
    p=10**y/(1.0+10**y)
    dp=math.log(10)*p*(1-p)*dy
    if t<1.0/Gamma:
        density=2*(N*mu/p)*math.exp(-c1*c3*N*p/t)*dp
    elif t>=1/Gamma:
        density=2*(N*mu/p)*math.exp(-c1*c3*N*Gamma*p)*dp
        
    return density

def transient_drift_draft_solution_middle_log(N, mu, Gamma, c1, c3, t, y, dy):
    
    p=10**y/(1.0+10**y)
    dp=math.log(10)*p*(1-p)*dy
    
    ss=-N*mu*dp*math.log(1.0-(1.0/(c1*c3*Gamma*N))/p)
    return ss
```

```python
#Theoretical solutions for regime 3

def mutation_draft_solution_peak_log(N, mu, Gamma, cc2, cc3, y, dy):
    
    p=10**y/(1.0+10**y)
    dp=math.log(10)*p*(1-p)*dy
    
    phii=0.4*cc2*mu/(cc3*Gamma)
    phii_bar=1.0-phii
    
    if p<0.5:
        density=(2.2/phi)*math.exp(-p/phii)*dp
    else:
        density=(2.2/phi)*math.exp(-(1-p)/phii)*dp

    return density

def mutation_draft_solution_low_log(N, mu, Gamma, cc2, cc3, y, dy):
    
    p=10**y/(1.0+10**y)
    dp=math.log(10)*p*(1-p)*dy
    
    phii=cc2*mu/(cc3*Gamma)
    phii_bar=1.0-phii
    
    density=-0.8*(1.0/phii)*(-(1-p/phii)*math.log((1-p/phii))+(p/phii)*(math.log(phii)-1))*dp

    return density

def mutation_draft_solution_high_log(N, mu, Gamma, cc2, cc3, y, dy):
    
    p=10**y/(1.0+10**y)
    dp=math.log(10)*p*(1-p)*dy
    
    phii=cc2*mu/(cc3*Gamma)
    phii_bar=1.0-phii
    
    density=-0.8*(1.0/phii)*(-(1-(1-p)/phii)*math.log((1-(1-p)/phii))+((1-p)/phii)*(math.log(phii)-1))*dp

    return density

def mutation_draft_solution_middle_log(N, m, G, cc2, cc3, y, dy):
    
    p=10**y/(1.0+10**y)
    dp=math.log(10)*p*(1-p)*dy
    
    phii=cc2*m/(cc3*G)
    phii_bar=1.0-phii

    density=(0.5*phii*(phii-phii_bar)/((p-phii)*(p-phii_bar)))*dp
    return density


def mutation_draft_solution_low_transient_log(N, m, G, cc2, cc3, t, y, dy):
    
    p=10**y/(1.0+10**y)
    dp=math.log(10)*p*(1-p)*dy
    
    c_t=0.08 #coefficient for how time relates to 1/Gamma
    phii=c_t*cc2*m*t/(cc3)
    phii_bar=1.0-phii
    
    density=-0.2*(1.0/phii)*(-(1-p/phii)*math.log((1-p/phii))+(p/phii)*(math.log(phii)-1))*dp
    return density

def mutation_draft_solution_middle_transient_log(N, m, G, cc2, cc3, t, y, dy):
    
    p=10**y/(1.0+10**y)
    dp=math.log(10)*p*(1-p)*dy
    
    c_t=0.09 #coefficient for how time relates to 1/Gamma
    
    phii=c_t*cc2*m*t/(cc3)
    phii_bar=1.0-phii
    
    density=c_t*(mu*t)*(G*t)*dp/(p-phii)
    
    return density

```

# Multiple sites

```python
# extract joint site frequencies from freq trajectories returns list for site 1 and list for site 2
def joint_SFS_at_t(freq_trajs, sampling_time, interval_start, interval_end):
    
    site_freqs_a=[]
    site_freqs_b=[]
    for traj in freq_trajs:
        
        for i, f in enumerate(traj):
            time_of_sample=(i*sampling_time)
            if interval_start<=time_of_sample<=interval_end:
                site_freqs_a.append(f[1]+f[3])
                site_freqs_b.append(f[2]+f[3])
    
    return [site_freqs_a, site_freqs_b]
```

```python
N=10**5
bins=10

miss_dict={}
for lf_m, lr_m in zip(log_freqs_miss, log_ranks_miss):
    log_freq_rounded=int(round(lf_m*bins))
    if log_freq_rounded in miss_dict:
        miss_dict[log_freq_rounded].append(lr_m)
    else:
        miss_dict[log_freq_rounded]=[lr_m]
        
syn_dict={}
for lf_m, lr_m in zip(log_freqs_syn, log_ranks_syn):
    log_freq_rounded=int(round(lf_m*20))
    if log_freq_rounded in syn_dict:
        syn_dict[log_freq_rounded].append(lr_m)
    else:
        syn_dict[log_freq_rounded]=[lr_m]
        
ratio_dict={}
for i in range(-5*bins, -1*bins):
    if (i in syn_dict) and (i in miss_dict):
        ratio_dict[i]=np.mean(syn_dict[i])-np.mean(miss_dict[i])
        
selective_effect_dict={}
k_list=list(ratio_dict.keys())
for i in range(len(k_list)-1):
    density=10**ratio_dict[k_list[i+1]]-10**ratio_dict[k_list[i]]
    s=1.0/(N*10**(float(k_list[i])/bins))
    selective_effect_dict[s]=density
```

Assuming only deleterious sites exist, the extent to which nsyn sites decline relative to syn sites over a small frequency range tells you how many sites were unable to climb about tht frequency due to selection so assuming a completely sharp cut-off in the deleterious SFS, you can simply back out the number of sites with selection coefficients in that range. I did this above and plot the density of sites with selective coefficients in a given selection bin. This has a roughly linear relation so that if $x=log(s)$ then $rho(x) dx = (c-kx)dx$ with $c\approx0.008$ and $k\approx0.016$. This means that the density of sites as a function of selective effect is $\rho(s) \sim -k \log(s)/s$
